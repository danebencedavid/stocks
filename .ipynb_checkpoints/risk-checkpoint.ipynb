# Project
In this project a few stocks (AMZN,TSLA,PFE,BABA) are selected to be analysed.

## Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import yfinance as yf
```

```python
#Downloading the data
multi_data = yf.download(["AMZN","TSLA","PFE","BABA"],start="2023-01-01",end="2024-06-01")
```

```python
print(type(multi_data))
```

```python
print(multi_data.head())
```

```python
#Reducing the number of columns by dropping some of them. The original dataframe is still intact, a new rmulti_data is created.
to_drop = ["Open", "High", "Low"]
rmulti_data = multi_data.drop(columns=to_drop)
```

```python
rmulti_data.info()
```

```python
#Plotting Close Prices for each stocks
close_prices = rmulti_data["Close"]
plt.figure(figsize=(14,7))
for stock in close_prices.columns:
    sns.lineplot(data=close_prices[stock], label=stock)
```

```python
#Calculating 50-day adn 200-day moving averages
moving_avgs = pd.DataFrame()
for stock in close_prices.columns:
    moving_avgs[f'{stock} 50-day MA'] = close_prices[stock].rolling(window=50).mean()
    moving_avgs[f'{stock} 200-day MA'] = close_prices[stock].rolling(window=200).mean()
```

```python
moving_avgs
```

```python
combined_data = pd.concat([close_prices, moving_avgs], axis=1)
```

```python
moving_avgs = combined_data.dropna()
```

```python
moving_avgs
```

```python
plt.figure(figsize=(14, 7))

for stock in close_prices.columns:
    sns.lineplot(data=close_prices[stock], label=f'{stock} Close')
    sns.lineplot(data=moving_avgs[f'{stock} 50-day MA'], label=f'{stock} 50-day MA', linestyle='--')
    sns.lineplot(data=moving_avgs[f'{stock} 200-day MA'], label=f'{stock} 200-day MA', linestyle=':')

plt.title('Closing Prices and Moving Averages from Jan 2023 to June 2024')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(title='Stocks and Moving Averages')
plt.show()
```

```python
rmulti_data
```

```python
#Adjusted Close resample to a monthly frequency to calculate a monthly return
monthly_adj_close_data = rmulti_data["Adj Close"].resample('ME').last()
monthly_adj_close_data
```

```python
monthly_adj_close_return = monthly_adj_close_data.pct_change()*100
monthly_adj_close_return = monthly_adj_close_return.dropna()
monthly_adj_close_return
```

```python
#Calculating the cumulative returns
daily_returns = rmulti_data['Adj Close'].pct_change() #daily percentage returns

cumulative_returns = (1 + daily_returns).cumprod() - 1
cumulative_returns = cumulative_returns.dropna()
cumulative_returns
```

```python
plt.figure(figsize=(16, 8))
for stock in cumulative_returns.columns:
    sns.lineplot(data=cumulative_returns[stock], label=f'{stock} Cumulative Return')
```

```python
plt.figure(figsize=(14, 7))
for stock in monthly_adj_close_return.columns:
    sns.barplot(x=monthly_adj_close_return.index, y=monthly_adj_close_return[stock], label=f'{stock} Monthly Return')

plt.xticks(rotation=45)
plt.show()
```

```python
#Plotting the volume of each stock
plt.figure(figsize=(14, 7))

for stock in rmulti_data['Volume'].columns:
    sns.lineplot(data=rmulti_data['Volume'][stock], label=f'{stock} Volume')

plt.title('Trading Volume from Jan 2023 to June 2024')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend(title='Stocks')
plt.xticks(rotation=45)
plt.show()
```

```python
#Identifying unusual high and low volume periods
volume_stat = rmulti_data["Volume"].rolling(window=20).agg(['mean','std'])
volume_stat = volume_stat.dropna()
volume_stat
```

```python
h_threshold = volume_stat.xs('mean', axis=1, level=1) + 2 * volume_stat.xs('std', axis=1, level=1)
l_threshold = volume_stat.xs('mean', axis=1, level=1) - 2 * volume_stat.xs('std', axis=1, level=1)
```

```python
aligned_high_threshold = h_threshold.reindex_like(rmulti_data['Volume'])
aligned_low_threshold = l_threshold.reindex_like(rmulti_data['Volume'])
```

```python
high_volume = rmulti_data['Volume'] > aligned_high_threshold
low_volume = rmulti_data['Volume'] < aligned_low_threshold
```

```python
plt.figure(figsize=(20, 10))

for stock in rmulti_data['Volume'].columns:
    sns.lineplot(data=rmulti_data['Volume'][stock], label=f'{stock} Volume')

    # Filter the indices where high_volume and low_volume are True
    high_volume_indices = high_volume[stock][high_volume[stock]].index
    low_volume_indices = low_volume[stock][low_volume[stock]].index

    plt.scatter(high_volume_indices, multi_data['Volume'][stock][high_volume[stock]], color='red', label=f'{stock} High Volume', marker='^', s=100)
    plt.scatter(low_volume_indices, multi_data['Volume'][stock][low_volume[stock]], color='green', label=f'{stock} Low Volume', marker='v', s=100)

plt.title('Trading Volume with Highlighted High and Low Volume Periods')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend(title='Volume')
plt.xticks(rotation=45)
plt.show()
```
